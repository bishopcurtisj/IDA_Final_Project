## Clean data from Spotify Song Charts

```python
songs = pd.read_csv('..\\data\\songs.csv')
songs
```

```python
songs['movement']=songs['rank']-songs['previous']
songs['streams']=songs['streams'].apply(lambda x: x.replace(',','')).astype(int)
```

```python
songs['year']=songs['year'].apply(lambda x: str(x))
songs['song'] = songs['song'].apply(lambda x: x.strip())
```

```python
for index, row in songs.iterrows():
    if row['song'].find('feat')!=-1:
        s=row['song']
        songs.at[index, 'song']=s[:s.find('feat')-1].strip()
        
        s = (s[s.find('(')+1:s.find(')')])
        if s.find('feat')==0:
            s = (s[6:])
            artists = re.split(',|&', s)
        else:
            s = s[5:]
            s, end = s.split(' feat. ')
            artists = re.split(',|&', s)
            artists.append(end)
        for artist in artists:
            songs = songs.append({'year':row['year'],'rank':row['rank'],'song':songs.loc[index, 'song'],'name':artist.strip(),'peak':row['peak'],'previous':row['previous'],'streak':row['streak'],'streams':row['streams'],'movement':row['movement']},ignore_index=True)
    if row['song'].find('(with')!=-1:
        s=row['song']
        songs.at[index, 'song']=s[:s.find('(with')-1].strip()
        
        s = (s[s.find('(')+1:s.find(')')])
        if s.find('with')==0:
            s = (s[5:])
            artists = re.split(',|&', s)
        
        for artist in artists:
            songs = songs.append({'year':row['year'],'rank':row['rank'],'song':songs.loc[index, 'song'],'name':artist.strip(),'peak':row['peak'],'previous':row['previous'],'streak':row['streak'],'streams':row['streams'],'movement':row['movement']},ignore_index=True)


```

```python
songs.to_csv('..\\data\\songs.csv',index=False)
```

```python
songs.sort_values(by=['year','rank'],inplace=True)
songs=songs.reset_index(drop=True)
songs
```

## Clean data from Hits daily double

```python
hdd_charts = pd.read_csv('..\\data\\hdd_charts.csv')
```

```python

for index, row in hdd_charts.iterrows():
    for col in hdd_charts.columns:
        if col != 'date' and row[col] is not None:
            s = str(row[col])
            while s.find('>')!=-1:
                s =s[s.find('>')+1:]
                s=s[:s.find('<')]

            if col == 'artist' or col == 'label':
                hdd_charts.at[index,col]=s.strip()
            else:
                hdd_charts.at[index,col]=s.strip().strip("%").replace(',','')
    
    if row['lw']=='--':
        hdd_charts.at[index,'lw']=row['rank']

    if row['change']=='--':
        hdd_charts.at[index,'change']=0
    if row['tea']=='--':
        hdd_charts.at[index,'tea']=0
        hdd_charts.at[index,'sea']=0



cols = ['total',  'albums', 'tea', 'sea', 'rank', 'lw']
for col in cols:
    hdd_charts[col] = hdd_charts[col].apply(lambda x: 0 if x is None or x=='nan' else int(x))

hdd_charts['change'] = hdd_charts['change'].apply(lambda x: 0 if x is None or x=='nan' is None else float(x))

hdd_charts['change']= hdd_charts['change']/100

hdd_charts['movement']= hdd_charts['lw']-hdd_charts['rank']

hdd_charts.to_csv('..\\data\\hdd_charts.csv',index=False)
```

```python
hdd_charts.rename(columns={'albums': 'album_sales'}, inplace=True)
tracks = []

for index, row in hdd_charts.iterrows():
    split = row['artist'].split('|')
    if len(split)>1:
        tracks.append(split[1].strip())
        hdd_charts.at[index,'artist']=split[0].strip()
    else:
        tracks.append(None)

    

hdd_charts['album']=tracks
hdd_charts
```

```python
hdd_charts.rename(columns={'albums': 'album_sales'}, inplace=True)
```

```python
temp = hdd_charts[hdd_charts['artist'].str.contains('&amp;')]
art = ['ABOVE ','BELLE ', 'NATHANIEL RATELIFF ', 'EARTH, WIND ', 'BOB MARLEY ']
for index, row in temp.iterrows():
    split = row['artist'].split('&amp;')
    if split[0] not in art:
        s=split[1]
        split = split[0].split(',')
        split.append(s)
        for i in range(len(split)):
            split[i]=split[i].strip()

        hdd_charts.at[index,'artist']=split[0]
        for i in range(1,len(split)):
            hdd_charts = hdd_charts.append({'date':row['date'],'rank':row['rank'],'lw':row['lw'],'change':row['change'],'artist':split[i],'album_sales':row['album_sales'],'total':row['total'],'tea':row['tea'],'sea':row['sea'],'movement':row['movement'],'album':row['album']},ignore_index=True)

        
    
```

```python
hdd_charts=pd.read_csv('..\\data\\hdd_copy.csv')
```

```python
hdd_charts['year'] = hdd_charts['date'].apply(lambda x: x[-4:])
for index, row in hdd_charts.iterrows():
    pre = hdd_charts[(hdd_charts['post']==0)&(hdd_charts['year']==row['date'][-4:])]
    post = hdd_charts[(hdd_charts['post']==1)&(hdd_charts['year']==row['date'][-4:])]
    if row['post']==1 and row['artist'] not in pre['artist'].values:
        t = row['total']
        change = row['change']
        last_total = pre[pre['rank']==50]['total'].values[0]
        if change>0:
            total = t/(1+change)
        elif change<0:
            total = t/(1-change)
        else:
            total=last_total
        hdd_charts = hdd_charts.append({'date':pre['date'].values[0],'rank':0,'lw':0,'change':0,'artist':row['artist'],'label':row['label'],'album_sales':0,'total':total,'tea':0,'sea':0,'movement':0,'album':row['album'],'post':0,'halftime':row['halftime']},ignore_index=True)
    elif row['post']==0 and row['artist'] not in post['artist'].values:
        last = post[post['rank']==50]
        hdd_charts = hdd_charts.append({'date':post['date'].values[0],'rank':0,'lw':0,'change':0,'artist':row['artist'],'label':row['label'],'album_sales':last['album_sales'].values[0]*0.9,'total':last['total'].values[0]*0.9,'tea':last['tea'].values[0]*0.9,'sea':last['sea'].values[0]*0.9,'movement':0,'album':row['album'],'post':1,'halftime':row['halftime']},ignore_index=True)


hdd_charts.drop(columns=['year'],inplace=True)
hdd_charts.sort_values(by=['date','rank'],inplace=True)
hdd_charts.reset_index(drop=True, inplace=True)
```

```python
diff = []
for index, row in hdd_charts.iterrows():
    if row['post']==0:
        diff.append(0)
    else:
        pre = hdd_charts[(hdd_charts['post']==0)&(hdd_charts['artist']==row['artist'])]
        diff.append(row['total']-pre['total'].values[0])

hdd_charts['total_change']=diff
```
